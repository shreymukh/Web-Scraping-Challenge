```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup
import requests
import urllib.parse
import pymongo
import pprint
import pandas as pd
```

```python
# SCRAPING https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars

# Set the executable path and initialize the chrome browser in splinter
executable_path = {"executable_path":"chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless=False)
browser.visit("https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars")
```

```python
url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
response = requests.get(url)
print(response.status_code)
```

```python
searchterm = urllib.parse.quote('img')
print(searchterm)
```

```python
# Convert the browser html to a soup object 
soup = browser.html
news_soup = BeautifulSoup(soup, 'html.parser')
#print(news_soup.prettify())
```

```python
link = news_soup.find("div", class_='img')
print(link)
```

```python
link_all = news_soup.find_all("div", class_="img")
print(link_all)
```

```python
for l in link_all:
      print(l.contents[1])
```

```python
images = news_soup.find_all('img')
#print(images)
```

```python
for image in images:
    #print image source
    print(image['src'])    
```

```python
for image in images:
    #print alternate text
    print(image['alt']) 
```

```python
link=[]
name=[]
nasa_link = "https://www.jpl.nasa.gov"

for image in images:
    new_image = link.append(nasa_link + image['src'])
    new_name = name.append(image['alt'])
```

```python
# Convert to Pandas dataframe and export in csv
df = pd.DataFrame({'Nasa_Image':link, 'Img_name': name}) 
df
```
