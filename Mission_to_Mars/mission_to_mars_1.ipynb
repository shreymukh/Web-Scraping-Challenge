```python
#JPL Mars Space Images - Featured Image

# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup
import requests
import urllib.parse
import pymongo
import pprint
import pandas as pd
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.mars_info
collection = db.mars
```

```python
# SCRAPING 
# Set the executable path and initialize the chrome browser in splinter
executable_path = {"executable_path":"chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless=False)
browser.visit("https://mars.nasa.gov/news")
```

```python
searchterm = urllib.parse.quote('content_title')
print(searchterm)
```

```python
url = 'https://mars.nasa.gov/news'
response = requests.get(url)
print(response.status_code)
```

```python
#response.text
```

```python
# Convert the browser html to a soup object 
soup = browser.html
news_soup = BeautifulSoup(soup, 'html.parser')
```

```python
#print(news_soup.prettify())
```

```python
article = news_soup.find("div", class_='list_text').text
news_title = news_soup.find("div", class_="content_title").text
datetime = news_soup.find('div', class_='list_date').text
news_p = news_soup.find("div", class_ ="article_teaser_body").text
print(news_title)
print(datetime)
print(article)
print(news_p)
```

```python
# Retrieve the parent divs for all articles
results = news_soup.find_all('div', class_='list_text')

# loop over results to get article data
for result in results:
     # scrape the article header 
    header = result.find('div', class_='list_date').text
    
    # scrape the article subheader
    subheader = result.find('div', class_='content_title').text
    
    # scrape the datetime
    datetime = result.find('div', class_='article_teaser_body').text
    

    # print article data
    print('-----------------')
    print(header)
    print(subheader)
    print(datetime)
    
    
    # Dictionary to be inserted into MongoDB
    post = {
    'header': header,
    'subheader': subheader,
    'date': datetime,
    }

    # Insert dictionary into MongoDB as a document
    collection.insert_one(post)
```

```python
# Display the MongoDB records created above
headers = db.headers.find()
for header in headers:
    print(header)
```

```python
news_all = news_soup.find_all("div",class_="content_title")
date_all = news_soup.find_all('div', class_='list_date')
article_all = news_soup.find_all("div", class_='list_text')
para_all = news_soup.find_all("div", class_ ="article_teaser_body")
print(news_all)
print(date_all)
print(article_all)
print(para_all)
```

```python
for para in para_all:
    print(para.text)
```

```python
#List to store name of the product
Date_time=[]
News_Title=[]
Para_graph=[]
    
for a in news_soup.find_all("div", class_='list_text'):
    date = a.find('div',attrs={'class': 'list_date'})
    title= a.find('div', attrs={'class':'content_title'})
    par =  a.find('div',attrs={'class': 'article_teaser_body'})
                                     
    Date_time.append(date.text)  
    News_Title.append(title.text)
    Para_graph.append(par.text)
```

```python
# Convert to Pandas dataframe and export in csv
df = pd.DataFrame({'Date':Date_time, 'News_header': News_Title, 'Para': Para_graph}) 
df.head()
```

```python
df.to_csv('Nasa_Mars.csv', index=False, encoding='utf-8')
```
